```python
import pandas as pd
from geopy.geocoders import Nominatim
from meteostat import Point, Monthly
from datetime import datetime
import time

# Your city data
data = {
    'City': ['Montreal', 'Calgary', 'Ottawa', 'Edmonton', 'Winnipeg',
             'Mississauga', 'Vancouver', 'Brampton', 'Hamilton', 'Surrey',
             'Quebec City', 'Halifax', 'Laval', 'London'],
    'Population': [1492520, 1070930, 850525, 826785, 625260,
                   608645, 591675, 538825, 478115, 474265,
                   468435, 374790, 363165, 353305]
}

cities = pd.DataFrame(data)

# Initialize geolocator
geolocator = Nominatim(user_agent="city_geocoder")

# Function to get lat/lon from city name
def get_lat_lon(city):
    try:
        location = geolocator.geocode(city + ", Canada")
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except:
        return None, None

# Add Latitude and Longitude columns
cities[['Latitude', 'Longitude']] = cities['City'].apply(get_lat_lon).apply(pd.Series)

# Meteostat date range for climate averages (latest full year)
start = datetime(2023, 1, 1)
end = datetime(2023, 12, 31)

# Function to get average May-Sept temp for each city
def get_avg_temp(lat, lon):
    if pd.isna(lat) or pd.isna(lon):
        return None
    try:
        location = Point(lat, lon)
        data = Monthly(location, start, end).fetch()
        may_to_sep = data.loc[f'{start.year}-05':f'{start.year}-09', 'tavg']
        if not may_to_sep.empty:
            return may_to_sep.mean()
        else:
            return None
    except:
        return None

# Because geopy and Meteostat can throttle requests, add delay
avg_temps = []
for idx, row in cities.iterrows():
    avg_temp = get_avg_temp(row['Latitude'], row['Longitude'])
    avg_temps.append(avg_temp)
    time.sleep(1)  # polite delay to avoid throttling

cities['Avg_May_Sept_Temp_C'] = avg_temps

# Categorize function
def categorize_temp(avg_temp):
    if avg_temp is None:
        return 'Data not available'
    elif avg_temp < 15:
        return "❄️ Cold"
    elif avg_temp < 18:
        return "🌥️ Mild"
    elif avg_temp < 22:
        return "🌤️ Good"
    else:
        return "☀️ Warm"

cities['Category'] = cities['Avg_May_Sept_Temp_C'].apply(categorize_temp)

# Display the final DataFrame
print(cities[['City', 'Population', 'Latitude', 'Longitude', 'Avg_May_Sept_Temp_C', 'Category']])
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# --- Step 1: Create DataFrames ---
# You already have df_population, sorted_df (age), and weather_df
# Let's assume the names are consistent and we merge on 'City'

# Example: rename your `sorted_df` to df_age
df_age = sorted_df

# Step 2: Merge all DataFrames on 'City'
merged_df = df_age.merge(weather_data, on='City', how='left').merge(population_df, on='City', how='left')
# Rename columns for clarity
merged_df.rename(columns={
    'Aged 15 to 34 Total': 'Youth_Population',
    'Avg_May_Sept_Temp_C': 'Avg_Temp',
    'Category': 'Weather_Category'
}, inplace=True)
```

```python

# Step 3: Normalize numerical values to 0–1 scale
scaler = MinMaxScaler()
merged_df[['Norm_Population', 'Norm_Youth_Population', 'Norm_Temp']] = scaler.fit_transform(
    merged_df[['Population_x', 'Youth_Population', 'Avg_Temp']]
)

# Step 4: Assign weights and compute a common score
# You can adjust weights as needed
merged_df['Score'] = (
    0.4 * merged_df['Norm_Youth_Population'] +
    0.4 * merged_df['Norm_Population'] +
    0.2 * merged_df['Norm_Temp']
)


merged_df
```

```python

# Step 5: Rank based on Score
merged_df['Rank'] = merged_df['Score'].rank(ascending=False).astype(int)

# Step 6: Sort by Rank
final_df = merged_df.sort_values(by='Rank')

# Optional: Clean up for display
final_df = final_df[['Rank', 'City', 'Population_x', 'Youth_Population', 'Avg_Temp', 'Weather_Category', 'Score']]
```
